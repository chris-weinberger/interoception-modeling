```python
import numpy as np
import pandas as pd

df = pd.read_csv('executive.dandrois', sep='\t',header=None)
df.columns=['patient ID', 'time', 'condition1', 'scan-in-condition1', 'condition2', 'condition3', 'ignore', 'brain','nan']
df=df.drop(columns=['nan'])
df.head()
```

```python
df.head()
```

```python
# first grab data that is executive rest for controls
exec_rest_controls = df.loc[(df['patient ID'] < 3000) & (df['patient ID'] >= 2000) & (df['condition2']== 5)]
exec_rest_controls.head()
```

```python
exec_rest_controls.to_csv('executive_rest_controls.csv')
```
