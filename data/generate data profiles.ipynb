# Individual patient criticism-rest

```python
# EXECUTIVE
# we want to isolate the data from a single patient. get 18 data points of criticism, 
# followed by 18 points of rest, then repeat that. 72 points total.

df = pd.read_csv('executive.dandrois', sep='\t',header=None)
df.columns=['patient ID', 'time', 'condition1', 'scan-in-condition1', 'condition2', 'condition3', 'ignore', 'brain','nan']
df=df.drop(columns=['nan'])

# grab all the patient IDs, so we have a list of unique patients
column_values = df[["patient ID"]].values.ravel()
patients =  pd.unique(column_values)

print(patients)

# for now, we can focus on control patients--get a single patient in 2000s
df = df.loc[(df['patient ID'] == 2303)]

# now filter for criticism and rest conditions
df = df.loc[(df['condition2'] != 1) & (df['condition2'] != 3)]

df = df.reset_index()

# get where criticism starts
idx = df[df['condition2']==2].index.min()
print("idx: " + str(idx))
df = df.iloc[idx:idx+72]

df=df.drop(columns=['index'])

# save to csv for this patient
df.to_csv('2303_executive_threat_76_scans.csv')

df.head()
```

# Save all control + patient neutral-rest data

The segment below will save all control + patient data with the following format:

18 scans neutral + 18 scans rest + 18 scans neutral + 18 scans rest

```python
# loop which will format and save data for all neutral-rest control patients

df = pd.read_csv('interoceptive_forward.dandrois', sep='\t',header=None)
df.columns=['patient ID', 'time', 'condition1', 'scan-in-condition1', 'condition2', 'condition3', 'ignore', 'brain','nan']
df=df.drop(columns=['nan'])

# grab all the patient IDs, so we have a list of unique patients
column_values = df[["patient ID"]].values.ravel()
patients =  pd.unique(column_values)

# filter patients array to ignore 1000s - these are pilots
filter_arr = [(item>2000) for item in patients]
patients = patients[filter_arr]

# regions that we will iterate through
regions = ['executive', 'interoceptive_forward', 'salience_forward']

for patient in patients:
    for region in regions:
        openfile = f'{region}.dandrois'
        df = pd.read_csv(openfile, sep='\t',header=None)
        df.columns=['patient ID', 'time', 'condition1', 'scan-in-condition1', 'condition2', 'condition3', 'ignore', 'brain','nan']
        df=df.drop(columns=['nan'])

        # for now, we can focus on control patients--get a single patient in 2000s
        df = df.loc[(df['patient ID'] == patient)]

        # now filter for neutral and rest conditions
        df = df.loc[(df['condition2'] != 1) & (df['condition2'] != 2)]

        df = df.reset_index()

        # get where neutral starts
        idx = df[df['condition2']==3].index.min()
        print("idx: " + str(idx))
        df = df.iloc[idx:idx+72]

        df=df.drop(columns=['index'])
        
        file_name = f'./neutral_rest_76_scans/{patient}_{region}_neutral_76_scans.csv'
        print(file_name)
#         df.to_csv(file_name)
        
    
```
